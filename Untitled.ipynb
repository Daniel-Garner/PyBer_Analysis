```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv(r"C:\Users\JamesGarner\Desktop\BootCamp\Module 5\PyBer_Analysis\resources\PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df = pyber_ride_df.rename(columns={'Month': 'Month', 'Avg. Fare ($USD)': 'Average Fare'})
```

```python
pyber_ride_df.plot(x="Month", y="Average Fare")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Average Fare")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Average Fare")
plt.show()
```
